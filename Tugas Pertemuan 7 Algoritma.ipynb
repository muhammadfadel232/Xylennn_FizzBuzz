```
# Fungsi rekursif
def fibonacci(N):
    if N <= 1:
        return N
    else:
        return fibonacci(N-2) + fibonacci(N-1)

# Fungsi rekursif buat M kali N
def m_kali_n(m, n):
    if n == 0:
        return 0
    else:
        return m + m_kali_n(m, n-1)

# Program utama
while True:
    print()
    print("Nim Genap")
    print("menu pilihan")
    print("1. Barisan Fibonacci")
    print("2. M Kali N")
    print("0. Keluar")

    pilih = input("pilih menu : ")
    print()

    if pilih == '1':
        suku = int(input("Masukkan Jumlah Suku : "))
        print("barisan fibonacci sebanyak", suku, "suku :")

        # Looping cetak fibonacci
        for i in range(1, suku + 1):
            if i == suku:
                print(fibonacci(i))
            else:
                print(fibonacci(i), end=", ")

    elif pilih == '2':
        bil1 = int(input("Masukkan Suatu Bilangan Bulat : "))
        bil2 = int(input("Masukkan Suatu Bilangan Pengali : "))
        print()

        hasil = m_kali_n(bil1, bil2)
        print(bil1, "x", bil2, "=", hasil)

    elif pilih == '0':
        break
```
